```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def y(t):
    return np.cos(t)
def y1(t):
    return -np.sin(t)
```

```python
t    = np.linspace(-3*np.pi,2*np.pi,1000)
orig = y1(t)
```

```python
h       = 1
result1 = []
result2 = []
result3 = []

for i in t:
    derivative = ( y(i+h) - y(i) ) / h
    result1.append(derivative)

for i in t:
    derivative = ( y(i+h/2) - y(i-h/2) ) / h
    result2.append(derivative)
    
for i in t:
    derivative = ( 8*( y(i+h/4) - y(i-h/4) ) - (y(i+h/2) - y(i-h/2) ) ) / (3*h)
    result3.append(derivative)


plt.figure(figsize=(16,9))
plt.plot(t, orig   , 'k', ls='-' , label='Original')
plt.plot(t, result1, 'g', ls='--', label='Forward Difference')
plt.plot(t, result2, 'b', ls='--', label='Central Difference')
plt.plot(t, result3, 'r', ls='--', label='Extended Difference')
plt.legend()
plt.title('h = {}'.format(h))
plt.show()
```
